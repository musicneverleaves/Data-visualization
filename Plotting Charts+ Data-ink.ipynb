```python
import pandas as pd
import numpy as np
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.plotting import figure, output_notebook, show
from math import pi

output_notebook()

x =['Victoria', 'Queensland', 'Tasmania', 'New South Wales', 'Western Australia', 'Southern Australia']
y = [463000, 617000, 302000, 104000, 99000, 140000]
p = figure(x_range=x, plot_height=350, title="Australian bushfires' affected areas",
           toolbar_location=None, tools="")

p.vbar(x=x, top=y, width=0.2, color="gray")
p.yaxis.visible=False

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.outline_line_color = None   
p.xaxis.major_label_orientation = pi/4
p.y_range.start = 0

source = ColumnDataSource(data=dict(xlabel= x,
                                   yvals = y,
                                   ylabel= y))
labels = LabelSet(x='xlabel', y='yvals', text='ylabel', level='glyph',
              x_offset=-17, y_offset=0, text_font_size="8pt", source=source, render_mode='css')
p.add_layout(labels)

show(p)
```

```python
import pandas as pd
import numpy as np
from bokeh.models import FactorRange
from bokeh.plotting import figure, output_file, show
output_file("Male and female in different subjects.html")

x1 = [79,77,82,83]
x2 = [81,78,80,84]
ylabels= ['English-Male','English-Female','Math-Male','Math-Female']

p = figure(y_range=FactorRange(*ylabels), plot_width=400, plot_height=400, title="Male and female in different subjects",
           toolbar_location=None, tools="")

p.circle(x1,y= ['English-Male','English-Female','Math-Male','Math-Female'], size=20, color="red", alpha=0.5)
p.circle(x2,y= ['English-Male','English-Female','Math-Male','Math-Female'], size=20, color="blue", alpha=0.5)
p.line([79,81], y= ['English-Male','English-Male'], line_width=2)
p.line([77,78], y= ['English-Female','English-Female'], line_width=2)
p.line([82,80], y= ['Math-Male','Math-Male'], line_width=2)
p.line([83,84], y= ['Math-Female','Math-Female'], line_width=2)
show(p)
```
